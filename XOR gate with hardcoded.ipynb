```python
# XOR Gate using Two Neurons (Perceptron)

import numpy as np
def step_activation(x):
    if x >= 0:
        return 1
    else:
        return 0
def neuron(inputs, weights, bias):
    inputs = np.array(inputs)
    weights = np.array(weights)
    weighted_sum = np.dot(inputs, weights) + bias
    output = step_activation(weighted_sum)
    return weighted_sum, output
w_h1 = [1, 1]
b_h1 = -0.5

w_h2 = [1, 1]
b_h2 = -1.5

w_o = [1, -2]
b_o = -0.5


dataset = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]
print("XOR Gate using Perceptron")
print("---------------------------------------")

print("\nTruth Table")
print("x1  x2  Output")
for data in dataset:
    _, h1 = neuron(data, w_h1, b_h1)
    _, h2 = neuron(data, w_h2, b_h2)

    _, result = neuron([h1, h2], w_o, b_o)

    print(data[0], " ", data[1], "      ", result)
```
