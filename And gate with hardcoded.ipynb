```python
# AND Gate using a Single Neuron (Perceptron)
import numpy as np
def step_activation(x):
    if x >= 0:
        return 1
    else:
        return 0

def neuron(inputs, weights, bias):
    inputs = np.array(inputs)
    weights = np.array(weights)

    weighted_sum = np.dot(inputs, weights) + bias
    output = step_activation(weighted_sum)
    return weighted_sum, output

weights = [1, 1]
bias = -1.5

dataset = [
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]
print("Deep Learning AND Gate using Perceptron")
print("---------------------------------------")
print("Weights:", weights)
print("Bias:", bias)
print("\nTruth Table")
print("x1  x2  Weighted Sum  Output")
for data in dataset:

    z, result = neuron(data, weights, bias)

    print(data[0], " ", data[1], "     ", z, "        ", result)
```
